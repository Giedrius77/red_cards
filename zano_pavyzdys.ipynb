```python
# PYTHON GAME
# BASIC LOGIC:
# YOU AND AI START WITH 3 CARDS, CARDS ARE PICKED AT RANDOM WITH A SCORE SYSTEM, THE CARDS THAT YOU'RE GIVEN CANNOT BE HIGHER THAN 20 AND NOT LESS THAN 5, SAME FOR THE AI, THEN YOU'RE GIVEN THE OPTION TO "HIT" OR "PASS", IF YOU "HIT", YOU'LL GET A RANDOM NUMBER OF CARDS ADDED TO YOUR DECK, FROM 2-10, IF YOU "PASS" NOTHING WILL HAPPEN AND YOU'LL GIVE UP THE TURN TO THE AI.
# THEN THE DEALER PLAYS. THEY TAKE CARDS UNTIL THEY HAVE AT LEAST 17 POINTS.
# IF THE POINTS ARE EQUAL ON 17, 18 OR 19
# THE DEALER WINS. EQUAL ON 20, 21 OR BLACKJACK NO ONE WINS.
# THE PLAYER CAN ASK FOR MORE CARDS. IF THEY GET HIGHER THAN 21, THEY LOSE.
# RULES:
# CARDS 2 - 10 CARRY A FACE VALUE CONSISTENT WITH THEIR NUMBER WHILE ANY FACE CARD (JACK, QUEEN, KING) IS VALUED AT 10
# AND THE ACE IS EITHER 1 OR 11, DEPENDING ON HOW THE PLAYER REPRESENTS IT.
# OBJECTIVE:
# THE DEALER IS THE ONLY OPPONENT OF THE PLAYER BEING THE SOLE TARGET TO BEAT 
```

```python
# pvz 1

import random

def draw_card():
    card = random.randint(2, 11)  # Ace can be 1 or 11, hence using 11 as max
    if card == 11:  # Assume Ace is 11 by default, can be changed to 1 later if needed
        return 'A'
    else:
        return card

def get_card_value(card, current_score):
    if card == 'A':
        return 1 if current_score + 11 > 21 else 11
    else:
        return card

def draw_initial_hand():
    hand = []
    score = 0
    while len(hand) < 3:
        card = draw_card()
        hand.append(card)
        score += get_card_value(card, score)
        if score > 20:
            return draw_initial_hand()  # Redraw if score exceeds 20
    return hand

def calculate_score(hand):
    score = 0
    aces_count = 0
    for card in hand:
        if card != 'A':
            score += get_card_value(card, score)
        else:
            aces_count += 1
    for _ in range(aces_count):  # Handle Aces at the end
        score += get_card_value('A', score)
    return score

def main():
    player_hand = draw_initial_hand()
    dealer_hand = draw_initial_hand()

    player_score = calculate_score(player_hand)
    dealer_score = calculate_score(dealer_hand)

    print(f"Your initial cards: {player_hand}, score: {player_score}")
    print(f"Dealer's initial card: {dealer_hand[0]}")

    # Player's turn
    while True:
        choice = input("Do you want to 'hit' or 'pass'? ").lower()
        if choice == 'hit':
            player_hand.append(draw_card())
            player_score = calculate_score(player_hand)
            print(f"Your cards: {player_hand}, score: {player_score}")
            if player_score > 21:
                print("You lost! Your score exceeded 21.")
                return
        elif choice == 'pass':
            break
        else:
            print("Invalid input. Please type 'hit' or 'pass'.")

    # Dealer's turn
    print(f"Dealer's initial cards: {dealer_hand}, score: {dealer_score}")
    while dealer_score < 17:
        dealer_hand.append(draw_card())
        dealer_score = calculate_score(dealer_hand)
        print(f"Dealer takes a card: {dealer_hand}, score: {dealer_score}")

    # Compare scores
    if dealer_score >= 17 and dealer_score <= 21:
        if player_score > dealer_score:
            print("You win!")
        elif player_score == dealer_score and dealer_score in (17, 18, 19):
            print("Dealer wins on a tie between 17, 18, or 19!")
        elif player_score == dealer_score and dealer_score in (20, 21):
            print("It's a tie with no winner on a score of 20 or 21!")
        else:
            print("Dealer wins!")

    # If dealer busts
    if dealer_score > 21:
        print("Dealer busts, you win!")

if __name__ == "__main__":
    main()
```
